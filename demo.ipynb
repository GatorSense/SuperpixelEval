```python
import numpy as np
from PIL import Image
from pysnic.algorithms.snic import snic
from skimage.color import rgb2lab
from skimage.segmentation import mark_boundaries
import matplotlib.pyplot as plt
from metrics import *
```

### Segment Alligator

```python
rgb_alligator = np.array(Image.open('./alligator.jpg'))
lab_alligator = rgb2lab(rgb_alligator)

K = 300
m = 1
n = rgb_alligator.shape[0]*rgb_alligator.shape[1]

labels, _, _ = snic(lab_alligator.tolist(), K, m)
labels = np.array(labels)

fig = plt.figure(figsize = (8,8))
plt.imshow(mark_boundaries(rgb_alligator, labels, color = (0,0,1)))
plt.axis('off')
plt.show()
```

### Run Metrics

```python
CH = CalinskiHarabasz(rgb_alligator, labels)
LocalCH = LocalCalinskiHarabasz(rgb_alligator, labels)
LocalCH2 = LocalCalinskiHarabasz2(rgb_alligator, labels)

C = Compactness(rgb_alligator, labels)

DB = DaviesBouldin(rgb_alligator, labels)
LocalDB = LocalDaviesBouldin(rgb_alligator, labels)

D = Dunn(rgb_alligator, labels)
LocalD = LocalDunn(rgb_alligator, labels)

I = I(rgb_alligator, labels)
LocalI = LocalI(rgb_alligator, labels)

RS = RSquared(rgb_alligator, labels)
LocalRS = LocalRSquared(rgb_alligator, labels)

RMS = RMSSTD(rgb_alligator, labels)
LocalRMS = LocalRMSSTD(rgb_alligator, labels)

V = Variance(rgb_alligator, labels)

XB = XieBeni(rgb_alligator, labels)
LocalXB = XieBeni(rgb_alligator, labels)
```
